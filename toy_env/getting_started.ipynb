```python
import random
import matplotlib.pyplot as plt
import numpy as np
from toy_env.grid import grid_env
```

```python
grid = grid_env(to_plot=True, grid=False)
s = grid.get_state()
for i in range(50):
    a = random.sample(range(4), 1)
    s, r, dead = grid.step(a)
    if not dead:
        grid.plot()
        #print(f"state: {s}")
    else:
        print("Died in step", i, ", restarting")
        s = grid.reset()
print(grid.get_state())
print("Finished")
plt.show(block=True)
```
